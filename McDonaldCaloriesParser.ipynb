```python
mcUrl = 'http://www1.staging.mcdonalds.com/gnia/ws/category/getCategoryDetails/jsonp'
params = {
    'ctry': 'tw',
    'lang': 'ch',
    'callback': 'displayCategoryProducts',
    'cat': '100002',
    'callback': 'jsonp1520173367522',
    '_': '1520173369403'
}

resp = requests.get(mcUrl, params=params)
print(resp.text)
```

```python
jsonText = resp.text[resp.text.find('(') + 1: resp.text.rfind(')')]
nutientJson = json.loads(jsonText)
print(json.dumps(nutientJson, indent=4, ensure_ascii=False))
```

```python
foods = nutientJson['category']['items']['item']
print('{},{},{},{},{}'.format('品項', '熱量', '蛋白質', '脂肪', '碳水化合物'))
for food in foods:
    name = food['name']
    nutient = food['nutrient_facts']['nutrient']
    protein = nutient[0]['value']
    fat = nutient[1]['value']
    carbs = nutient[2]['value']
    kcal = nutient[4]['value']
    print('{},{},{},{},{}'.format(name, kcal, protein, fat, carbs))
```
